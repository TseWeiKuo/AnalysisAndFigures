# Ablation Landing-Probability Figures

Focused notebook for the ablation and no-contact landing-probability panels currently defined in `Seaborn_plot.py`.

## 1. Environment and output setup

```python
%matplotlib inline

from pathlib import Path
import sys

import matplotlib.colors as mcolors
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

ROOT = Path.cwd()
if not (ROOT / "Seaborn_plot.py").exists():
    ROOT = ROOT.parent

DATA_FOLDER = Path(r"C:\Users\agrawal-admin\Desktop\LandingDataSummary")
OUTPUT_FOLDER = ROOT / "Figures" / "AblationFigure"
OUTPUT_FOLDER.mkdir(parents=True, exist_ok=True)

TRIAL_NUM = 20
TRIAL_OFFSET = 3
LATENCY_THRESHOLD_S = 0.71

print(f"Python executable: {sys.executable}")
print(f"Data folder: {DATA_FOLDER}")
print(f"Output folder: {OUTPUT_FOLDER}")
```

## 2. Landing-probability helpers

These helpers reproduce the current `Seaborn_plot.py` rules: exclude the first three trials, classify landing latency below 0.71 s as successful, and treat `-1` or latency at/above 0.71 s as failed.

```python
def read_landing_probability(file_path, group_name, fps, first_fly, last_fly):
    raw = pd.read_excel(file_path)
    raw = raw.iloc[first_fly - 1:last_fly].copy()

    # Match the original threshold-based fly filtering.
    valid_rows = []
    for index, row in raw.iterrows():
        missing_count = sum(isinstance(value, str) or pd.isna(value) for value in row.values)
        if missing_count < len(row.values) / 2:
            valid_rows.append(index)
    raw = raw.loc[valid_rows]

    trial_columns = [f"Trial_{trial}" for trial in range(TRIAL_OFFSET + 1, TRIAL_NUM + 1)]
    missing_columns = [column for column in trial_columns if column not in raw.columns]
    if missing_columns:
        raise ValueError(f"{file_path} is missing trial columns: {missing_columns}")

    rows = []
    for row_index, row in raw[trial_columns].iterrows():
        successful = []
        failed = []
        for value in row:
            if isinstance(value, str) or pd.isna(value):
                continue
            value = float(value)
            if 0 < value < LATENCY_THRESHOLD_S * fps:
                successful.append(value / fps)
            elif value == -1 or value >= LATENCY_THRESHOLD_S * fps:
                failed.append(value)

        denominator = len(successful) + len(failed)
        if denominator == 0:
            continue
        rows.append({
            "Fly#": int(row_index) + 1,
            "LandingProb": len(successful) / denominator,
            "MLandingLatency": np.mean(successful) if successful else np.nan,
            "Group_Name": group_name,
            "n_valid_trials": denominator,
            "n_success": len(successful),
            "n_failed": len(failed),
        })

    return pd.DataFrame(rows)


def plot_landing_probability_panel(
        dataframes,
        file_name,
        colors,
        markers=None,
        box_width=0.25,
        box_softness=0.65,
):
    if markers is None:
        markers = ["o"] * len(dataframes)

    combined = pd.concat(dataframes, ignore_index=True)
    group_order = [frame["Group_Name"].iloc[0] for frame in dataframes]
    x_positions = np.arange(len(group_order))
    fig, ax = plt.subplots(figsize=(max(7, 2 * len(group_order)), 7))

    def soften(color):
        rgb = np.asarray(mcolors.to_rgb(color), dtype=float)
        return tuple(rgb + (1 - rgb) * np.clip(box_softness, 0, 1))

    rng = np.random.default_rng(0)
    for i, (frame, color, marker) in enumerate(zip(dataframes, colors, markers)):
        values = frame["LandingProb"].dropna().to_numpy(dtype=float)
        if len(values) == 0:
            continue

        box = ax.boxplot(
            [values],
            positions=[i - 0.20],
            widths=box_width,
            patch_artist=True,
            showfliers=False,
            boxprops=dict(facecolor=soften(color), edgecolor=color, linewidth=2),
            whiskerprops=dict(color=color, linewidth=2),
            capprops=dict(color=color, linewidth=2),
            medianprops=dict(color="black", linewidth=2),
        )

        jittered_x = rng.normal(i + 0.08, 0.025, size=len(values))
        ax.scatter(
            jittered_x,
            values,
            color=color,
            marker=marker,
            s=90,
            alpha=0.5,
            zorder=5,
        )

    stat_df = (
        combined.groupby("Group_Name", sort=False)["LandingProb"]
        .agg(["mean", "std", "count"])
        .reset_index()
    )

    ax.set_xticks(x_positions)
    ax.set_xticklabels(group_order, rotation=45, ha="right")
    ax.set_ylim(-0.1, 1.1)
    ax.set_yticks([0, 0.5, 1])
    ax.set_ylabel("Landing probability")
    ax.set_xlim(-0.5, len(group_order) - 0.5)
    sns.despine(ax=ax)
    fig.tight_layout()

    output_prefix = OUTPUT_FOLDER / file_name
    fig.savefig(f"{output_prefix}.pdf", dpi=300, bbox_inches="tight")
    stat_df.to_csv(f"{output_prefix}-stat.csv", index=False)
    display(stat_df)
    plt.show()
    return fig, ax, combined, stat_df
```

## 3. Load ablation and no-contact datasets

```python
ablation_specs = [
    {
        "label": "WT-SL-T1-CxTr",
        "path": DATA_FOLDER / "Necessity" / "T1RightIntact_CTF_LL_All.xlsx",
        "fps": 300, "first_fly": 1, "last_fly": 21,
    },
    {
        "label": "WT-SL-T2-TiTa",
        "path": DATA_FOLDER / "Necessity" / "T2RightIntact_TTa_All.xlsx",
        "fps": 300, "first_fly": 1, "last_fly": 21,
    },
    {
        "label": "WT-SL-T2-CxTr",
        "path": DATA_FOLDER / "Necessity" / "T2RightIntact_CTF_LL_All.xlsx",
        "fps": 300, "first_fly": 1, "last_fly": 20,
    },
    {
        "label": "WT-SL-T3-TiTa",
        "path": DATA_FOLDER / "Necessity" / "T3RightIntact_TTa_All.xlsx",
        "fps": 300, "first_fly": 1, "last_fly": 17,
    },
    {
        "label": "WT-SL-T3-CxTr",
        "path": DATA_FOLDER / "Necessity" / "T3RightIntact_CTF_LL_All.xlsx",
        "fps": 300, "first_fly": 1, "last_fly": 20,
    },
]

no_contact_specs = [
    {
        "label": "WT Abdomen",
        "path": DATA_FOLDER / "CONTROL" / "WT-Abdomen-ALL.xlsx",
        "fps": 250, "first_fly": 1, "last_fly": 11,
    },
    {
        "label": "WT Ab T3CutOff",
        "path": DATA_FOLDER / "CONTROL" / "WT-Ab-LegCutOff-ALL.xlsx",
        "fps": 250, "first_fly": 1, "last_fly": 10,
    },
    {
        "label": "No contact",
        "path": DATA_FOLDER / "CONTROL" / "IntactFly_Control.xlsx",
        "fps": 300, "first_fly": 1, "last_fly": 10,
    },
]

def load_specs(specs):
    datasets = []
    for spec in specs:
        if not spec["path"].exists():
            raise FileNotFoundError(spec["path"])
        datasets.append(read_landing_probability(
            spec["path"], spec["label"], spec["fps"],
            spec["first_fly"], spec["last_fly"],
        ))
    return datasets

print("Ablation paths:")
for spec in ablation_specs:
    print(spec["path"], spec["path"].exists())
print("\nNo-contact paths:")
for spec in no_contact_specs:
    print(spec["path"], spec["path"].exists())
```

## 4. Ablation landing-probability panel

```python
RUN_ABLATION_LP = False

if RUN_ABLATION_LP:
    ablation_data = load_specs(ablation_specs)
    ABLATION_COLORS = ["red", "green", "blue", "red", "green"]
    plot_landing_probability_panel(
        dataframes=ablation_data,
        file_name="Ablation",
        colors=ABLATION_COLORS,
        markers=["o"] * len(ablation_data),
        box_width=0.28,
        box_softness=0.70,
    )
```

## 5. No-contact/control landing-probability panel

```python
RUN_NO_CONTACT_LP = False

if RUN_NO_CONTACT_LP:
    no_contact_data = load_specs(no_contact_specs)
    NO_CONTACT_COLORS = ["red", "green", "blue"]
    plot_landing_probability_panel(
        dataframes=no_contact_data,
        file_name="NoContact",
        colors=NO_CONTACT_COLORS,
        markers=["o"] * len(no_contact_data),
        box_width=0.28,
        box_softness=0.70,
    )
```

## Usage notes

1. Update `DATA_FOLDER` if the landing-data summary folder moves.
2. Run Sections 1-3 first.
3. Set only `RUN_ABLATION_LP` or `RUN_NO_CONTACT_LP` to `True`.
4. Figures and summary CSV files are saved under `Figures/AblationFigure`.
5. The soft box fill and its outline use the same color as each group’s stripplot points.
